```python
import sagemaker
from sagemaker import get_execution_role
```

```python
sagemaker_session = sagemaker.Session()
```

```python
role = get_execution_role()
```

```python
role
```

```python
train_input = sagemaker_session.upload_data('data')
```

```python
train_input
```

```python
from sagemaker.sklearn.estimator import SKLearn

script_path = "Lineer_Reg.py"
sklearn = SKLearn(
entry_point=script_path, instance_type="ml.m4.xlarge",
framework_version="0.20.0", py_version="py3",
role=role,
    sagemaker_session=sagemaker_session)
```

```python
sklearn.fit({"train": train_input})
```

```python
deployment = sklearn.deploy(initial_instance_count=1, instance_type="ml.m4.xlarge")
```

```python
sample = [[7.4, 0.70, 0.00, 1.9, 0.076, 11.0, 34.0, 0.9978, 3.51, 0.56, 9.4]]

prediction = deployment.predict(sample)
print(prediction)
```
